```python
#import scraping and API libraries
from bs4 import BeautifulSoup
import requests
print("imported BeautifulSoup and requests successfully!")
```

```python
#open wikipedia page and scrape html script
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml') #lxml is the html parser  
# print(soup.prettify()) #prints html script of entire wikipedia page, prettify() method is passed to make output easier to read
```

```python
table = soup.find('table',{"class":"wikitable sortable"}).tbody #finds html table body content inside of <table> tag
print(table.prettify()) #prints html script of <table> tag, prettify() method is passed to make output easier to read
```

```python
rows = table.find_all('tr') #extracts all table row content embedded within <table> tags and assigns it to variable "rows"
rows
```

```python
#Create empty arrays to split the table contents with respect to the columns (postal code, boroughs, neighborhoods)
postal_code=[]
boroughs=[]
neighborhoods=[]

#Use for loop to systematically extract data from table rows
for i in range(1, len(rows)): #starts from 2nd row (index = 1, excludes headers)
    tds = rows[i].find_all('td') #pass find_all method to extract all table data from rows and assign it to variable "tds"
#     print(tds) #uncomment to view raw table data
    values = [td.text.replace('\n','') for td in tds] #extracts clean version of table data values from raw table data and assigns it to variable "values"
    postalcode = values[0] #extracts postal code data [first item (index = 0) in "values" table data] and assigns it to variable "postalcode"  
    borough = values[1] #extracts borough data [second item (index = 1) in "values" table data] and assigns it to variable "borough"
    neighborhood = values[2] #extracts neighborhood data [third item (index = 2) in "values" table data] and assigns it to variable "neighborhood"
    #append data to lists
    postal_code.append(postalcode)
    boroughs.append(borough)
    neighborhoods.append(neighborhood)
```

```python
columns = ['PostalCode','Borough','Neighborhood'] #create list of column headers

toronto_df = pd.DataFrame(list(zip(postal_code, boroughs, neighborhoods)), columns=columns) #create pandas df and append columns list to it
toronto_df.head()
```

<h3> Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned. </h3>

```python
toronto_df["Borough"].replace("Not assigned", np.nan, inplace=True) #replace cells with "Not assigned" with NaN values in "Borough" column
toronto_df.head()
```

```python
toronto_df.dropna(subset=["Borough"], axis=0, inplace=True) #delete cells with "Not assigned" values (i.e. NaN) in "Borough" column
toronto_df.head()
```

<h3>Combine neighborhoods that exist in the same postal code area.</h3>

```python
toronto_df=toronto_df.groupby("PostalCode").agg(lambda x:','.join(set(x)))
toronto_df.head()
```

```python
toronto_df = toronto_df.reset_index() #resets index column in df
toronto_df.head()
```

```python
toronto_df.to_csv('toronto.csv',index=False) #export df to csv file
```

## Explore Neighborhoods in Downtown Toronto
#### Let's create a function to repeat the same process to all the neighborhoods in Downtown Toronto

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

#### Write the code to run the above function on each neighborhood and create a new dataframe called *dt_toronto_venues*.

```python
dt_toronto_venues = getNearbyVenues(names=dt_toronto_data['Neighborhood'],
                                   latitudes=dt_toronto_data['Latitude'],
                                   longitudes=dt_toronto_data['Longitude']
                                  )
```

## Analyze Each Neighborhood

```python
# one hot encoding
dt_toronto_onehot = pd.get_dummies(dt_toronto_venues[['Venue Category']], prefix="", prefix_sep="")
dt_toronto_onehot.head()
```

```python
# add neighborhood column back to dataframe
dt_toronto_onehot['Neighborhood'] = dt_toronto_venues['Neighborhood'] 
dt_toronto_onehot.head()
```

```python
# move neighborhood column to the first column
fixed_columns = [dt_toronto_onehot.columns[-1]] + list(dt_toronto_onehot.columns[:-1])
fixed_columns
dt_toronto_onehot = dt_toronto_onehot[fixed_columns]
dt_toronto_onehot.head()
```

```python
dt_toronto_onehot.shape
```

#### Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
dt_toronto_grouped = dt_toronto_onehot.groupby('Neighborhood').mean().reset_index()
dt_toronto_grouped.head()
```

```python
dt_toronto_grouped.shape
```

#### Let's print each neighborhood along with the top 5 most common venues

```python
num_top_venues = 5

for hood in dt_toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = dt_toronto_grouped[dt_toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:] #to help convert strings to float.
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### Let's add this data into a Pandas Dataframe

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#Let's create the new dataframe and display the top 10 venues for each neighborhood.

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = dt_toronto_grouped['Neighborhood']

for ind in np.arange(dt_toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(dt_toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Cluster Neighborhoods
Run *k*-means to cluster the neighborhood into 5 clusters.

```python
# set number of clusters
kclusters = 5

dt_toronto_grouped_clustering = dt_toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dt_toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
#Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

dt_toronto_merged = dt_toronto_data

# merge newyork_grouped with newyork_data to add latitude/longitude for each neighborhood
dt_toronto_merged = dt_toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

dt_toronto_merged # check the last columns!
```

```python
#Visualize resulting clusters

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=13)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(dt_toronto_merged['Latitude'], dt_toronto_merged['Longitude'], 
                                  dt_toronto_merged['Neighborhood'], dt_toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster+1), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
